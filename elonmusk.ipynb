# Classification

```python
import pandas as pd
from transformers import AutoTokenizer,AutoModelForSequenceClassification
from scipy.special import softmax


#Load model and tokenizer
roberta = "cardiffnlp/twitter-roberta-base-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)


# Définir la longueur maximale du tweet
max_tweet_length = 128  # Vous pouvez ajuster cette valeur en fonction des exigences du modèle

# Prétraitement et analyse de sentiment pour chaque tweet
for tweet_text in tweet:
    # Prétraitement du tweet et troncation
    tweet_words = []
    for word in tweet_text.split(' '):
        if word.startswith('@') and len(word) > 1:
            word = '@user'
        elif word.startswith('http'):
            word = "http"
        tweet_words.append(word)
    tweet_proc = " ".join(tweet_words)[:max_tweet_length]  # Tronquer le tweet au maximum de la longueur acceptée

    # Analyse de sentiment
    encoded_tweet = tokenizer(tweet_proc, return_tensors='pt', truncation=True, max_length=max_tweet_length)
    output = model(**encoded_tweet)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)

    # Stockage des résultats dans un nouveau DataFrame
    results_df = pd.DataFrame({
        'Tweet': [tweet_text],
        'Negative': [scores[0]],
        'Neutral': [scores[1]],
        'Positive': [scores[2]]
    })

    # Enregistrement des résultats dans un fichier CSV
    with open('sentiment_elonmusk_results.csv', 'a', newline='', encoding='utf-8') as csvfile:
        results_df.to_csv(csvfile, header=not csvfile.tell(), index=False, encoding='utf-8')
```
